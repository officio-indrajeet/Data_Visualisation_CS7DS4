```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('gapminder.csv')
data
```

```python
filtered_data = data[(data['year'] >= 1957) & (data['year'] <= 2007)]
filtered_data
```

```python
plt.figure(figsize=(12, 8))
x_range = list(range(1957,2008,5))
plt.xticks(x_range)
x = filtered_data['year'] + np.random.uniform(-1,1,size = len(filtered_data))
sns.scatterplot(x=x, y='lifeExp', hue='continent', data=filtered_data, palette='Set1')
plt.title('Life Expectancy Over the Years (1957-2007) - Color Encoding')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend(title='Continent')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
x_range = list(range(1957,2008,5))
plt.xticks(x_range)
x = filtered_data['year'] + np.random.uniform(-1,1,size = len(filtered_data))
sns.scatterplot(x=x, y='lifeExp', style='continent', markers=['o', 's', 'D', '^', 'v', 'P'], data=filtered_data)
plt.title('Life Expectancy Over the Years (1957-2007) - Shape Encoding')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend(title='Continent')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
x_range = list(range(1957,2008,5))
plt.xticks(x_range)
x = filtered_data['year'] + np.random.uniform(-1.5,1.5,size = len(filtered_data))
sns.scatterplot(x=x, y='lifeExp', size='continent', sizes=(30, 150), data=filtered_data)
plt.title('Life Expectancy Over the Years (1957-2007) - Size Encoding')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
filtered_data = data[['gdpPercap', 'lifeExp', 'pop']]
filtered_data
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdpPercap', y='lifeExp', hue='pop', data=filtered_data, palette='viridis')
plt.title('Correlation Between GDP, Life Expectancy, and Population - Brightness Encoding')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdpPercap', y='lifeExp', size='pop', data=filtered_data, sizes=(100, 500))
plt.title('Correlation Between GDP, Life Expectancy, and Population - Position Encoding')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
textures = ['/', '\\', '|', '-', '+']
filtered_data['texture_index'] = pd.cut(filtered_data['pop'], bins=len(textures), labels=textures)

# Create the scatterplot with texture encoding
plt.figure(figsize=(10, 6))
x = filtered_data['gdpPercap'] + np.random.uniform(-2000, 2000, size=len(filtered_data))
sns.scatterplot(x=x, y='lifeExp', style='texture_index', data=filtered_data)
plt.title('Life Expectancy vs. GDP per Capita (Texture Encoding)')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.tight_layout()
plt.show()
```

```python
colors = {
    'Africa': 'red',
    'Asia': 'blue',
    'Europe': 'green',
    'Americas': 'orange',
    'Oceania': 'purple'
}

marker_shapes = {
    'Low': 'o',  # Circle
    'Medium': 's',  # Square
    'High': '^',  # Triangle
}

data['pop_scaled'] = (data['pop'] - data['pop'].min()) / (data['pop'].max() - data['pop'].min()) * 500

data['gdp_category'] = pd.cut(data['gdpPercap'], bins=[0, 10000, 20000, float('inf')], labels=['Low', 'Medium', 'High'])

plt.figure(figsize=(10, 6))

for i, row in data.iterrows():
    jittered_year = row['year'] + np.random.uniform(-1, 1)
    plt.scatter(jittered_year, row['lifeExp'], s=row['pop_scaled'], c=colors[row['continent']],
                marker=marker_shapes[row['gdp_category']], label=row['continent'])

legend_elements = [
    plt.Line2D([0], [0], marker='o', color='b', label='Low GDP', markerfacecolor='b', markersize=5),
    plt.Line2D([0], [0], marker='s', color='b', label='Medium GDP', markerfacecolor='b', markersize=5),
    plt.Line2D([0], [0], marker='^', color='b', label='High GDP', markerfacecolor='b', markersize=5),
    plt.Line2D([0], [0], marker='.',color='red', label='Africa', markersize=10, linestyle='None'),
    plt.Line2D([0], [0], marker='.',color='blue', label='Asia', markersize=10, linestyle='None'),
    plt.Line2D([0], [0], marker='.',color='green', label='Europe', markersize=10, linestyle='None'),
    plt.Line2D([0], [0], marker='.',color='orange', label='Americas', markersize=10, linestyle='None'),
    plt.Line2D([0], [0], marker='.',color='purple', label='Oceania', markersize=10, linestyle='None'),
]

plt.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Scatterplot of Multiple Attributes : LifeExpectancy, Year, Continent, Population and GDP ')

# Add grid
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
